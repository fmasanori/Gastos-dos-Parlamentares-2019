```python
import pandas as pd
```

```python
low_memory=False
%matplotlib inline
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90
```

```python
filename = '/Users/iagosaito/Dados Públicos/Ano-2019.csv'
```

```python
columns = ['txNomeParlamentar', 
                   'ideCadastro', 
                   'nuCarteiraParlamentar',
                   'nuLegislatura', 
                   'sgUF', 
                   'sgPartido', 
                   'codLegislatura', 
                   'numSubCota',
                   'txtDescricao', 
                   'numEspecificacaoSubCota', 
                   'txtDescricaoEspecificacao', 
                   'txtFornecedor',
                   'txtCNPJCPF', 
                   'txtNumero', 
                   'indTipoDocumento',
                   'datEmissao', 
                   'vlrDocumento', 
                   'vlrGlosa', 
                   'vlrLiquido', 
                   'numMes', 
                   'numAno', 
                   'numParcela', 
                   'txtPassageiro', 
                   'txtTrecho',
                   'numLote', 
                   'numRessarcimento', 
                   'vlrRestituicao', 
                   'nuDeputadoId', 
                   'ideDocumento']
df = pd.DataFrame(pd.read_csv(filename, engine="python", sep=';', names = columns))
```

```python
df.shape
```

```python
df.rename(columns={'txNomeParlamentar': 'NOME_PARLAMENTAR', 
                   'ideCadastro': 'ID_CADASTRO', 
                   'nuCarteiraParlamentar': 'NUMERO_CARTEIRA_PARLAMENTAR',
                   'nuLegislatura': 'NUMERO_LEGISLATURA', 
                   'sgUF': 'UF', 
                   'sgPartido': 'PARTIDO', 
                   'codLegislatura': 'COD_LEGISLATURA', 
                   'numSubCota': 'NUMERO_SUB_COTA',
                   'txtDescricao': 'DESCRICAO', 
                   'numEspecificacaoSubCota': 'NUMERO_ESPECIFICACAO_SUB_COTA', 
                   'txtDescricaoEspecificacao': 'DESCRICAO_ESPECIFICACAO', 
                   'txtFornecedor': 'FORNECEDOR',
                   'txtCNPJCPF': 'CPF_CNPJ', 
                   'txtNumero': 'NUMERO', 
                   'indTipoDocumento': 'IND_TIPO_DOCUMENTO',
                   'datEmissao': 'DATA_EMISSAO', 
                   'vlrDocumento': 'VALOR_DOCUMENTO', 
                   'vlrGlosa': 'VALOR_GLOSA', 
                   'vlrLiquido': 'VALOR_LIQUIDO', 
                   'numMes': 'NUMERO_MES', 
                   'numAno': 'ANO', 
                   'numParcela': 'NUMERO_PARCELA', 
                   'txtPassageiro': 'PASSAGEIRO', 
                   'txtTrecho': 'TRECHO',
                   'numLote': 'NUMERO_LOTE', 
                   'numRessarcimento': 'NUMERO_RESSASRCIMENTO', 
                   'vlrRestituicao': 'VALOR_RESTITUICAO', 
                   'nuDeputadoId': 'ID_DEPUTADO', 
                   'ideDocumento': 'ID_DOCUMENTO'}, inplace=True)
```

```python
#Tipos de gastos
df.groupby('DESCRICAO').size().sort_values().tail(20).plot(kind='barh', figsize=(10,10))
```

```python
#Estados que mais registraram gastos
df.groupby('UF').size().sort_values().tail(20).plot(kind='barh', figsize=(10,10))
```

```python
# Parlamentares que mais registraram gastos
df.groupby('NOME_PARLAMENTAR').size().sort_values().tail(20).plot(kind='barh', figsize=(10,10))
```

```python
#Partidos que mais registraram gastos
df.groupby('PARTIDO').size().sort_values().tail(20).plot(kind='barh', figsize=(10,10))
```

```python
#Valor Gasto por Partido

df.groupby('PARTIDO').agg({'VALOR_LIQUIDO': 'sum'}).sort_values(by='VALOR_LIQUIDO').tail(40).plot(kind='barh', figsize=(10,10))
```

```python
#Valor gasto por estado

df.groupby('UF').agg({'VALOR_LIQUIDO': 'sum'}).sort_values(by='VALOR_LIQUIDO').tail(40).plot(kind='barh', figsize=(10,10))
```

```python
#Valor gasto por parlamentar

df.groupby('NOME_PARLAMENTAR').agg({'VALOR_LIQUIDO': 'sum'}).sort_values(by='VALOR_LIQUIDO').tail(30).plot(kind='barh', figsize=(10,10))
```
